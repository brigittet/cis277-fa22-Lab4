## NFL Offense Data

1. Import the NFL week 3, 4, and 5 data and add a week column to dataset and set the value to 3, 4, or 5 for all rows from that week.

```python
import pandas as pd
df_3 = pd.read_csv("nfloffenseweek3.csv")
df_4 = pd.read_csv("nfloffenseweek4.csv")
df_5 = pd.read_csv("nfloffenseweek5.csv")
df_3["week"] = 3
df_4["week"] = 4
df_5["week"] = 5
```

2. Combine the weekly dataframes by adding the dataframes together row-wise.

```python
df_3.merge(df_4, how = "outer").merge(df_5, how = "outer")
```

3. Make up 1 interesting question about this dataset. Answer it.

How do the top 10 teams in terms of total yards compare between week 3 and week 5?

```python
df_all = df_3.merge(df_4, how = "outer").merge(df_5, how = "outer")
week_3_most_yards = df_all[df_all.week == 3].sort_values("total_yards", ascending = False).head(10).team
print("Week 3:")
print(week_3_most_yards)
week_5_most_yards = df_all[df_all.week == 5].sort_values("total_yards", ascending = False).head(10).team
print("Week 5:")
print(week_5_most_yards)
```

9. Make up 1 interesting question about one of these datasets and answer it.

How does the average health of countries with population densities in the 75th or higher percentile compare with the average health of all countries?

```python
pql_df = pop_df.merge(qol_df, how = "outer")
high_density = pql_df[pql_df.pop_per_km_sq >= pql_df.pop_per_km_sq.quantile(0.75)]
print("Average Health of Countries with High Population Density: ", high_density.health.mean())
print("Average Health of All Countries", pql_df.health.mean())
```

10. Make up 1 interesting question about one of these datasets and answer it.

Is the average safety of countries higher in countries with above average daily max temperatures or below average daily max temperatures?

```python
iqqol_df = iq_df.merge(qol_df)
avg_temp = iqqol_df.daily_max_temp.mean()
above_avg = iqqol_df[iqqol_df.daily_max_temp > avg_temp]
below_avg = iqqol_df[iqqol_df.daily_max_temp < avg_temp]
print("Safety of Countries with Higher Temps:", above_avg.safety.mean())
print("Safety of Countries with Lower Temps:", below_avg.safety.mean())
```
